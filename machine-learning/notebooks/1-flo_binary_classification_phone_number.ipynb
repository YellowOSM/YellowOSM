```python
DIM = 50
from keras.models import Sequential
from keras.layers import Dense
```

```python
import random
import json
import numpy as np
import pandas as pd
```

```python
data = []
with open("training_data_1.json", 'r') as myfile:
    data = json.loads(myfile.read())
len(data)
```

```python
not_phones = []
phones = []

for el in data:
    phones.append([el[0]['phone_web'], 1])
    for not_phone in el[1]:
        not_phones.append([not_phone, 0])


training_data = np.zeros((len(phones)*2, 51))
        
# pvs = []
for p,i in zip(phones, range(0,len(phones))):
#     pv = []
    for c,j in zip(p[0], range(0,len(p[0]))):
        training_data[i,j] = ord(c)
#     pvs.append(pv + [int(p[1])])

random.shuffle(not_phones)
# npvs = []
for noph,i in zip(not_phones[:len(phones)], range(len(phones),(len(phones)*2))):
#     npv = []
    for c,j in zip(noph[0], range(0,len(noph[0]))):
        training_data[i,j] = ord(c)
#     npvs.append(npv + [int(noph[1])])


np.random.shuffle(training_data)
# training_data = pvs + npvs
# random.shuffle(training_data)
# # print("len phones: "+ str(len(phones))+ " " + str(len(training_data)))
```

```python
print(len(training_data), len(phones))
```

```python
training_data = training_data.astype(int)
```

```python
my_training_data = np.array(training_data)
x_train = my_training_data[:int(len(training_data)*0.9), :DIM]
y_train = my_training_data[:int(len(training_data)*0.9), DIM:]
x_test = my_training_data[int(len(training_data)*0.9):, :DIM]
y_test = my_training_data[int(len(training_data)*0.9):, DIM:]
```

```python
my_training_data.shape
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# from keras.utils import to_categorical
from keras.utils.np_utils import to_categorical
y_train_cat = to_categorical(y_train, num_classes=2, dtype='int32')
y_train_cat
y_test_cat = to_categorical(y_test, num_classes=2, dtype='int32')
```

```python
model = Sequential()
model.add(Dense(units=DIM, activation='relu', input_dim=DIM))
model.add(Dense(units=20, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
model.fit(x_train, y_train_cat, epochs=5, batch_size=320)
```

```python
loss_and_metrics = model.evaluate(x_test, y_test_cat, batch_size=128)
```
